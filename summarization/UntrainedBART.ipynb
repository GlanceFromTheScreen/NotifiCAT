```python
import pandas as pd
from simpletransformers.seq2seq import Seq2SeqModel,Seq2SeqArgs
import pickle
import torch
```

```python
model_args = Seq2SeqArgs()
model_args.num_train_epochs = 10
model_args.no_save = False
model_args.evaluate_generated_text = True
model_args.evaluate_during_training = False
model_args.evaluate_during_training_verbose = False
model_args.overwrite_output_dir = True
# model_args.temperature = 0.1
model_args.max_length = 400
model_args.max_seq_length = 400
model_args.min_seq_length = 100
model_args.best_model_dir = "best_model"


# Initialize model
model = Seq2SeqModel(
    encoder_decoder_type="bart",
    encoder_decoder_name="facebook/bart-large-cnn",
    args=model_args,
    use_cuda=False,
)
```

```python
torch.save(model, "model_clean_save_torch")
```

```python
input_file = 'serialized_variable_test.pkl'
with open(input_file, 'rb') as f:
    # Десериализуем переменную из файла
    deserialized_variable_test = pickle.load(f)

test_df = pd.DataFrame(
    deserialized_variable_test, columns=["input_text", "target_text"]
)
```

```python
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.bleu_score import SmoothingFunction

def calculate_rouge_score(hypothesis, reference):
    # Convert hypothesis and reference to lists of tokens
    hypothesis_tokens = hypothesis.split()
    reference_tokens = reference.split()

    # Calculate ROUGE score using overlap of bigrams
    bigram_overlap = set(hypothesis_tokens) & set(reference_tokens)
    rouge_score = len(bigram_overlap) / len(reference_tokens)

    return rouge_score

def calculate_bleu_score(hypothesis, reference):
    # Convert hypothesis and reference to lists of tokens
    hypothesis_tokens = hypothesis.split()
    reference_tokens = reference.split()

    # Calculate BLEU score with smoothing to handle short hypotheses
    smoothing = SmoothingFunction().method1
    bleu_score = sentence_bleu([reference_tokens], hypothesis_tokens, smoothing_function=smoothing)

    return bleu_score

# Example usage
hypothesis = "The cat sat on the mat."
reference = "The cat is on the mat."
rouge_score = calculate_rouge_score(hypothesis, reference)
bleu_score = calculate_bleu_score(hypothesis, reference)

print("ROUGE score:", rouge_score)
print("BLEU score:", bleu_score)
```

```python
mean_rouge = 0
mean_bleu = 0
for item in deserialized_variable_test:
  prediction = model.predict([item[0]])[0]
  mean_rouge += calculate_rouge_score(prediction, item[1])
  mean_bleu += calculate_bleu_score(prediction, item[1])
mean_rouge = mean_rouge / len(deserialized_variable_test)
mean_bleu = mean_bleu / len(deserialized_variable_test)

print('ROUGE: ', mean_rouge)
print('BLUE: ', mean_bleu)
```
